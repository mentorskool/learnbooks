```python
import boto3
import json
```

```python
# get dynamodb resource

dynamodb_res  = boto3.resource("dynamodb")
```

```python
"""
Evaluation
SkillMaster
"""

# get evaluation instance
evaluation_table = dynamodb_res.Table("Evaluation")
```

```python
# get all the data dump from the table

all_data = evaluation_table.scan()

# Items

all_data = all_data["Items"]
```
